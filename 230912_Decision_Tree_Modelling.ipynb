```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/230907_basic_table_for_analysis_cleaned.csv", index_col=0, low_memory=False)

df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df.date, format='%Y-%m-%d')
df['year'] = df['date'].dt.year.astype(int)

df.info()
```

```python
df.shape
```

```python
#drop columns not relevant for model

df.drop(['adr', 'lat', 'long', 'holiday','date'], axis=1, inplace=True)
```

```python
df.shape
df.head()
```

```python
#if cell before doesn't work

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['num_veh'] = le.fit_transform(df['num_veh'].astype(str))
df['actp'] = le.fit_transform(df['actp'].astype(str))
df['hrmn'] = le.fit_transform(df['hrmn'].astype(str))
```

```python
df.head()
```

```python
#separating into test and training data: 

#for training data keep years 2015-2019 
train = df[df['year'] < 2020]

#for test data keep years 2020-2021
test = df[df['year'] >= 2020]

```

```python
#define X_train, X_test, y_train, y_test

X_train = train.drop(['fatal'], axis=1)
y_train = train['fatal']

X_test = test.drop(['fatal'], axis=1)
y_test = test['fatal']
```

```python
#DecisionTree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV


# Tuning the parameters to find the best decision tree model - start params chosen randomly
model = DecisionTreeClassifier(criterion = 'gini', random_state=50)

param_dist = {'criterion': ['gini', 'entropy'],
              'max_depth': [2, 3, 4], 
              'min_samples_leaf' : [2,3,4]}

cv_dt = GridSearchCV(model, param_grid=param_dist, n_jobs=-1)

cv_dt.fit(X_train, y_train)
print('Best Parameters using grid search: \n', cv_dt.best_params_)

```

```python
# Tuning the parameters to find the best decision tree model - start params chosen randomly
model = DecisionTreeClassifier(random_state=50)

param_dist = {'criterion': ['gini', 'entropy'],
              'max_depth': [2,5,6,7], 
              'min_samples_leaf' : [2,5,6,7]}

cv_dt = GridSearchCV(model, param_grid=param_dist, n_jobs=-1)

cv_dt.fit(X_train, y_train)
print('Best Parameters using grid search: \n', cv_dt.best_params_)
```

```python
# Training with the best parameters

tree = DecisionTreeClassifier(criterion='gini', random_state = 50, max_depth=2, min_samples_leaf=2)
tree.fit(X_train, y_train)
y_pred_gini  = tree.predict(X_test)

print('Accuracy Score: ',accuracy_score(y_test, y_pred_gini))
```

```python
pd.crosstab(y_test, y_pred_gini, rownames=['Reality'], colnames=['Prediction'])
```

```python
# classification_report

print('Classification Report')
print(classification_report(y_test, y_pred_gini))

```

```python
#Training set accuracy

y_pred_train_gini = tree.predict(X_train)

y_pred_train_gini

print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(tree.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(tree.score(X_test, y_test)))
```

Here, the training-set accuracy score is 1.0000 while the test-set accuracy to be 0.9997. These two values are quite comparable. So, there is no sign of overfitting.
