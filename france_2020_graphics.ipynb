```python
# merge files
merged_df = pd.merge(df, df1, on='Num_Acc',suffixes=('_df', '_df1'))
merged_df = pd.merge(merged_df, df3, on='Num_Acc',suffixes=('_merged_df', '_df3'))
merged_df = pd.merge(merged_df, df4, on='Num_Acc',suffixes=('_merged_df', '_df4'))
```

```python
merged_df.head(5)
merged_df.info()
```

```python
duplicates = merged_df.duplicated()

# Count the number of duplicates
num_duplicates = duplicates.sum()

# Print the number of duplicates
print("Number of duplicates:", num_duplicates)
```

```python
event_counts = merged_df['lum_df'].value_counts()

# Plot the event counts
event_counts.plot.bar()
labels = ['Full day', 'Twilight or dawn', 'Night without public lighting', 'Night with public lighting not lit', 'Night with public lighting on','misinformation']
# Add labels and title to the plot
plt.xticks(range(len(event_counts)), labels)
plt.xlabel('Event')
plt.ylabel('Count')
plt.title('light conditions ')

# Display the plot
plt.show()
```

```python
event_counts = merged_df['col_df'].value_counts()

# Plot the event counts
event_counts.plot.bar()
labels = ['2 Frontal', '2 Rear', '2 by side', '3 or more chain', '3 + multiple colisions','others', 'without collision','error']
# Add labels and title to the plot
plt.xticks(range(len(event_counts)), labels)
plt.xlabel('Event')
plt.ylabel('Count')
plt.title('colision types ')

# Display the plot
plt.show()
```
