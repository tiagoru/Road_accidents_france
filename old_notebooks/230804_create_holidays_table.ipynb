```python
import pandas as pd
import numpy as np
import holidays
from datetime import date
```

# Creation of a holidays table
it might be useful to know about the holidays as an additional criterion for accidents.

- import the data table
- create a time object and add it to the merged table
- get the holidays and add it to the merged table

```python
merged_table = pd.read_csv('./data/230819_input_table.csv', low_memory=False, header = 0, index_col=0, na_values='n/a')
```

```python
# store date in string format
merged_table['date'] = merged_table.apply(lambda row: date(row['an'], row['mois'], row['jour']).strftime('%Y-%m-%d'), axis = 1)
merged_table.head()
```

```python
# get holiday list for France
fr_holidays = holidays.FR()
```

```python
merged_table['is_holiday'] = merged_table.apply(lambda row: row['date'] in fr_holidays, axis = 1)
```

```python
merged_table['holiday'] = merged_table.apply(lambda row: fr_holidays.get(row['date']) if row['is_holiday'] == True else np.nan, axis = 1)
```

```python
merged_table['date'] = pd.to_datetime(merged_table['date'])
merged_table.info()
```
