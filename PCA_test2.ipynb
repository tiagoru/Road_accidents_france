```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import mutual_info_regression , mutual_info_classif
```

```python
df = pd.read_csv("./data/231018_clean_table_for_analysis.csv")
```

```python
df= df.drop(["date","is_holiday","holiday"], axis =1)
import numpy as np
```

```python
# create target variable
# create a new variable for the severity ['Grav'] of accidents ( Sereve, non-severe) as a possible target variable
#### 0 = non severe and 1 equal to severe
df['Severe'] = np.where(df['grav'].isin([0, 1]), 0, 1)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assuming 'df' contains your data

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Perform PCA with 28 components
pca = PCA(n_components=28)
PC = pca.fit_transform(X_scaled)

# Create a DataFrame for the principal components
pc_df = pd.DataFrame(data=PC, columns=[f'principal component {i}' for i in range(1, 29)])

# Display the first few rows of the DataFrame
pc_df.head()
```

```python
Y_l = df['Severe'].tolist()
joined = pc_df
joined['Severe'] = Y_l
joined.head()
```

```python
import seaborn as sns

# Include 'Severe' for coloring the data points
pca_df = joined  # Use your DataFrame containing the PCA results

# Pairplot for all 29 principal components
sns.pairplot(pca_df, hue='Severe', diag_kind='kde', vars=[f'principal component {i}' for i in range(1, 29)])
plt.show()
```
