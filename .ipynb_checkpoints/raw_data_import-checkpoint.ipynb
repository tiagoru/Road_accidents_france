```python
import glob
import pandas as pd
```

```python
# read user files in data directory
files = []
for file in glob.glob("./data_import/usagers*.csv"):
    files.append(file)

# import user files
users = pd.DataFrame()

for i in range(0, len(files)):
    file = pd.read_csv(files[i], sep = ',')
    if i == 0:
        users = file
    else:
        users = pd.concat([users, file], axis = 0)

# export csv
users.to_csv("./data/users.csv", sep = ',', header = True, na_rep = 'n/a')
users.info()
```

```python
# read characteristics (=features) files in data directory
files = []
for file in glob.glob("./data_import/caracter*.csv"):
    files.append(file)

# import user files
features = pd.DataFrame()

for i in range(0, len(files)):
    file = pd.read_csv(files[i], sep = ',', encoding='latin-1', na_values=['N/A', 'n/a',''])
    if i == 0:
        features = file
    else:
        features = pd.concat([features, file], axis = 0)

# export csv
features.to_csv("./data/features.csv", sep = ',', header = True, na_rep = 'n/a')
features.info()
```

```python
# read places files in data directory
files = []
for file in glob.glob("./data_import/lieux*.csv"):
    files.append(file)

# import user files
places = pd.DataFrame()

for i in range(0, len(files)):
    file = pd.read_csv(files[i], sep = ',', encoding='latin-1', na_values=['N/A', 'n/a',''], low_memory=False)
    if i == 0:
        places = file
    else:
        places = pd.concat([features, file], axis = 0)

# export csv
places.to_csv("./data/places.csv", sep = ',', header = True, na_rep = 'N/A')
places.info()
```

```python
# read vehicle files in data directory
files = []
for file in glob.glob("./data_import/vehicules*.csv"):
    files.append(file)

# import user files
vehicles = pd.DataFrame()

for i in range(0, len(files)):
    file = pd.read_csv(files[i], sep = ',', encoding='latin-1', na_values=['N/A', 'n/a',''], low_memory=False)
    if i == 0:
        vehicles = file
    else:
        vehicles = pd.concat([vehicles, file], axis = 0)

# export csv
vehicles.to_csv("./data/vehicles.csv", sep = ',', header = True, na_rep = 'N/A')
vehicles.info()
```

```python
# read registered vehicle files in data directory
files = []
for file in glob.glob("./data_import/registered_vehicles*.csv"):
    files.append(file)

# import user files
vehicles = pd.DataFrame()

for i in range(0, len(files)):
    file = pd.read_csv(files[i], sep = ';', na_values=['N/A', 'n/a',''], low_memory=False, encoding="UTF-8")
    if i == 0:
        vehicles = file
    else:
        vehicles = pd.concat([vehicles, file], axis = 0)

vehicles.head()
```

```python
vehicles.info()
```

```python
# data wrangling
# Id_accident contains whitespace, needs to be removed

vehicles['Id_accident'] = vehicles['Id_accident'].replace(" ", "")
```

```python
vehicles['Id_accident']
```

```python
# The whitespace is not removed. Very strange.
# Let's check the string:
vehicles.iloc[0].Id_accident
```

```python
# well, this is unbreakable space...
# strange encoding..
# found this https://stackoverflow.com/questions/10993612/how-to-remove-xa0-from-string-in-python
```

```python
new_id = list()
for i in range(0, len(vehicles)):
    new_id.append(vehicles.iloc[i].Id_accident.replace('\xa0', ''))
```

```python
# new_id is still string, so let's change:
print(new_id[0])

new_id_numeric =pd.to_numeric(new_id)
print(new_id_numeric[0])
```

```python
vehicles.Id_accident = new_id_numeric
```
