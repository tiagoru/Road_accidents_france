```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# df = pd.read_csv("C:/Users/tiago/OneDrive/Documents/GitHub/mar23cds_accidents/data/features.csv", index_col=0)
df = pd.read_csv("./data/features.csv", index_col=0)

df.info()
df.head(5)
```

```python
# df1 = pd.read_csv('C:/Users/tiago/OneDrive/Documents/GitHub/mar23cds_accidents/data/places.csv', index_col=0)
df1 = pd.read_csv('./data/places.csv', index_col=0)
df1.info()
df1.head(5)
```

```python
# df2 = pd.read_csv('C:/Users/tiago/OneDrive/Documents/GitHub/mar23cds_accidents/data/registered_vehicles.csv', index_col=0)
df2 = pd.read_csv('./data/registered_vehicles.csv', index_col=0)
df2.info()
df2.head(5)
```

```python
# df3 = pd.read_csv('C:/Users/tiago/OneDrive/Documents/GitHub/mar23cds_accidents/data/users.csv', index_col=0)
df3 = pd.read_csv('./data/users.csv', index_col=0)

df3.info()
df3.head()
```

```python
# df4 = pd.read_csv('C:/Users/tiago/OneDrive/Documents/GitHub/mar23cds_accidents/data/vehicles.csv', index_col=0)
df4 = pd.read_csv('./data/vehicles.csv', index_col=0)
df4.info()
df4.head(5)
```

```python
# merge files
merged_df = pd.merge(df, df1, on='Num_Acc',suffixes=('_df', '_df1'))
merged_df = pd.merge(merged_df, df3, on='Num_Acc',suffixes=('_merged_df', '_df3'))
merged_df = pd.merge(merged_df, df4, on='Num_Acc',suffixes=('_merged_df', '_df4'))
```

```python
merged_df.head(5)
merged_df.info()
```

```python
merged_df.to_csv("./data/merged_tables.csv", sep = ',', header = True, na_rep = 'n/a', index=True)
```

```python
duplicates = merged_df.duplicated()

# Count the number of duplicates
num_duplicates = duplicates.sum()

# Print the number of duplicates
print("Number of duplicates:", num_duplicates)
```

```python
# After update, most duplicate variables disappeared
# event_counts = merged_df['lum_df'].value_counts()
event_counts = merged_df['lum'].value_counts()

# Plot the event counts
event_counts.plot.bar()
labels = ['Full day', 'Twilight or dawn', 'Night without public lighting', 'Night with public lighting not lit', 'Night with public lighting on','misinformation']
# Add labels and title to the plot
plt.xticks(range(len(event_counts)), labels)
plt.xlabel('Event')
plt.ylabel('Count')
plt.title('light conditions ')

# Display the plot
plt.show()
```

```python
# After update, most duplicate variables disappeared
# event_counts = merged_df['col_df'].value_counts()
event_counts = merged_df['col'].value_counts()

# Plot the event counts
event_counts.plot.bar()
labels = ['2 Frontal', '2 Rear', '2 by side', '3 or more chain', '3 + multiple colisions','others', 'without collision','error']
# Add labels and title to the plot
plt.xticks(range(len(event_counts)), labels)
plt.xlabel('Event')
plt.ylabel('Count')
plt.title('colision types ')

# Display the plot
plt.show()
```
